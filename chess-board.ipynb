```python
import numpy as np
import matplotlib.pyplot as plt

class ChessBoard:
    def __init__(self):
        
        self.board=np.zeros([8,8,3])
        self.white=np.ones([8,8,3])
        self.red=[]
        self.blue=[]
        
        """ another Method for chessboard """
        #         def build_checkerboard(w, h) :
        #       re = np.r_[ w*[0,1] ]              # even-numbered rows
        #       ro = np.r_[ w*[1,0] ]              # odd-numbered rows
        #       return np.row_stack(h*(re, ro))
        # checkerboard = build_checkerboard(4,4) #64 blocks
        # plt.imshow(checkerboard,cmap='gray')
        
        #########################################
       
        for i in range(0,8,2):
            for y in range (0,8,2):
                self.board[i][y] = self.white[i][i]
        for i in range(1,8,2):
            for y in range (1,8,2):
                self.board[i][y] = self.white[i][i]
    def render(self):
        """renders The chess Board"""
       
        plt.imshow(self.board)
        
    def add_red(self,row,col):
        """adds red queen"""
        red=[1,2,0]
        self.red=[row,col]
        self.board[row][col]=red
        
    def add_blue(self,row,col):
        """adds blue queen"""
        blue=[0,1,1]
        self.blue=[row,col]
        self.board[row][col]=blue
    def is_under_attack(self):
        
        """
        or we can make 3 methods and pass them as arguments for this function 
        """
        x1 , y1 = self.blue
        x2 , y2 = self.red
        x2_x1=abs(x2-x1)
        y2_y1=abs(y2-y1)
        
        if x2_x1 == y2_y1 or x2_x1 ==0 or y2_y1==0 :
            return True
        return False
    
    def reset_board(self):
        "resets the blacks and whites of chess board"
        self.board=np.zeros([8,8,3])
        for i in range(0,8,2):
            for y in range (0,8,2):
                self.board[i][y] = self.white[i][i]
        for i in range(1,8,2):
            for y in range (1,8,2):
                self.board[i][y] = self.white[i][i]
        
        
        
        

plot=ChessBoard()
plot.reset_board()
plot.add_blue(3,3)
plot.add_red(4,4)
plot.render()
plot.is_under_attack()

```

Diagonal Attack  --> True !

```python
def build_checkerboard(w, h) :
    
      """
      another method for printing the chess board !
      """

      re = np.r_[ w*[0,1] ]              # even-numbered rows
      ro = np.r_[ w*[1,0] ]              # odd-numbered rows
      return np.row_stack(h*(re, ro))


checkerboard = build_checkerboard(4,4) #64 blocks 4^4 = 64 :D

plt.imshow(checkerboard,cmap='gray')  
```
